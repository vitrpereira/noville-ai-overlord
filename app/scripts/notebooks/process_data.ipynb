```python
import pandas as pd
from datetime import timedelta
```

```python
df = pd.read_csv('/Users/vitorpereira/dev/noville/noville-ai-overlord/app/scripts/notebooks/transactions_data/jan_2025_expenses.csv')
```

```python
df.columns = df.rename(columns=str.lower).columns
```

```python
df.head()
```

```python
df.date = pd.to_datetime(df.date, dayfirst=True)

df
```

```python
columns_rename_mapping = {
    'fato gerador': 'description', 
    'parcelas': 'installments', 
    'valor total (r$)': 'amount', 
    'método': 'payment_method', 
    'operaçao': 'operation',
    'número da parcela': 'installment_number'
}

df.columns = df.rename(columns=columns_rename_mapping).columns

df
```

```python
df.operation = df.operation.str.lower()

df.operation = df.operation.replace({'crédito': 'credit', 'débito': 'debit'})

df.head()
```

```python
# Fetch all dates where installments and installment number > 1
# Say the date is the date minus the number of installment_number - 1 months back

df.date = df.apply(lambda row: row['date'] - pd.DateOffset(months=row['installment_number'] - 1) if row['installments'] > 1 and row['installment_number'] > 1 else row['date'], axis=1)

df.head()
```

```python
columns_to_drop = ['valor mês (r$)', 'agents', 'fracionado', 'installment_number']

df = df.drop(columns=columns_to_drop)
```

```python
df.payment_method = df.payment_method.str.lower()
df.head()
```

```python
df['user_id'] = 1

df
```

```python
# Transform amount into a float
df.amount = df.amount.str.replace('.', '').str.replace(',', '.').astype(float)
```

```python
df = df.sort_values(by='date')

df
```

```python
df.to_csv('/Users/vitorpereira/dev/noville/noville-ai-overlord/app/scripts/ingestion_data/transactions_data_jan_2025.csv', index=False)
```
